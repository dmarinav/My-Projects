```python
corr = df.iloc[:,1:].corr().round(decimals=2)
sns.set(font_scale=1.4)
plt.subplots( figsize=(13, 7))
#cmap = sns.diverging_palette(300, 350, as_cmap=True)
#cmap = sns.diverging_palette(65, 300, as_cmap=True)
cmap = sns.diverging_palette(85, 1, as_cmap=True)
sns.heatmap (corr, annot=True,alpha=.9, cmap=cmap,cbar_kws={"shrink": .9},linewidths=.5)
# vmax=.2
plt.show()
```

```python
df1=df.copy()

df1['Survived']=df['Survived'].apply(lambda x: 'No' if x ==0 else 'Yes')

df1['Sex']=df['Sex'].apply(lambda x: 'Female' if x =='female' else 'Male')

df1['Class']=df['Pclass']
```

```python
def embark(words):
        if 'Q' in str(words):
            return 'Queenstown'
        elif'C' in str(words):
            return 'Cherbourg'
        else:
            return "Southampton"
```

```python
df1['Embarked']=df["Embarked"].map(embark);


df1['Embarked'].value_counts().nunique();

df1['Embarked'].value_counts()
```

```python

sns.set(font_scale=1.2)
plt.style.use(['bmh'])

df2 = df[['Age', 'Fare', 'Parch','SibSp']];
df2.columns = ['Age', 'Passenger Fare', 'Parents/Children Aboard', 'Siblings/Spouses Aboard']

df2.hist(color='darkolivegreen',figsize=(12,8), ylabelsize =13, xlabelsize=13, alpha=.8)
plt.show()
```
